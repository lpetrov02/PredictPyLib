# Условие  
Прибор для выявления брака на фабрике имеет вероятность ошибки 5% (и первого и второго рода), процент брака составляет 5% от всего объёма выпускаемой продукции.  

#### 1. Какая вероятность того, что мы выявили брак, если прибор выдал положительный результат - "продукция бракованная"?

#### 2. Почему же в жизни все-таки используют такие приборы? Что можно было бы изменить в процедуре поиска брака, не меняя точности прибора, так, чтобы вероятность из первого вопроса P(брак|"+") выросла?

#### 3. Какое соотношение можно вывести между процентом брака P(брак) и ошибкой прибора, если мы хотим , чтобы прибор работал лучше честной монетки, хуже или также?

# Решение

#### (1)  
Обозначим следующие события:  
- **B** - продукт бракованный. $p_0 = \mathbb{P}(B)$ = 0.05 - вероятность брака  
- **NB** - продукт не бракованный  
- **+** - тест дал положительный результат, то есть сказал, что продукт бракованный  
- **-** - тест дал отрицательный результат  
Тогда одно из условий можно записать как $\mathbb{P}(+|NB) = \mathbb{P}(-|B)$ = 0.05  
В данном пункте требуется найти $\mathbb{P}(B|+)$  
Применим формулу Байеса, а потом - формулу полной вероятности:  
$$
\mathbb{P}(B|+) = \frac{\mathbb{P}(+|B)\mathbb{P}(B)}{\mathbb{P}(+)} = \frac{\mathbb{P}(+|B)\mathbb{P}(B)}{\mathbb{P}(+|B)\mathbb{P}(B) + \mathbb{P}(+|NB)\mathbb{P}(NB)}
$$  
Тут мы всё знаем, можно считать  
$$
\mathbb{P}(B|+) = \frac{(1 - 0.05) \cdot 0.05}{(1 - 0.05) \cdot 0.05 + 0.05 \cdot (1 - 0.05)} = \frac{1}{2}
$$

#### (2)  
Почему же такими приборами пользуются?  
Во-первых, какая-то ошибка есть почти у любого прибора, она просто может быть побольше или поменьше, но она всё равно, скорее всего, будет. А ничего лучше, чем "прибор с ошибкой" не придумали)
И, к тому же, есть способы увеличить вероятность из п.1.  
Рассмотрим обратную вероятность: $\mathbb{P}(NB|+)$.  
Проведём *N* независимых экспериентов. У нас появятся новые события:  
- $+_N$ - все *N* тест показал "брак"  
- $-_N$ - брак хотя бы раз не был обнаружен  
Эти события взаимно исключающие, поэтому будем выносить вердикт "брак", если $+_N$, и вердикт "не брак", если $-_N$.  
$$
\mathbb{P}(NB|+_N) = \frac{\mathbb{P}(+_N|NB)\mathbb{P}(NB)}{\mathbb{P}(+_N)} = \frac{\mathbb{P}(+_N|NB)\mathbb{P}(NB)}{\mathbb{P}(+_N|NB)\mathbb{P}(NB) + \mathbb{P}(+_N|B)\mathbb{P}(B)} = \frac{\mathbb{P}(+|NB)^N\mathbb{P}(NB)}{\mathbb{P}(+|NB)^N\mathbb{P}(NB) + \mathbb{P}(+|B)^N\mathbb{P}(B)} = \frac{0.05^N \cdot 0.95}{0.05^N \cdot 0.95 + 0.95^N \cdot 0.05} = \frac{1}{1 + 19^{N-1}} \to 0
$$  
Следовательно  
$$
\mathbb{P}(B|+_N) = 1 - \mathbb{P}(NB|+_N) \to 1
$$  
Таки вот образом мы можем увеличить вероятноть того, что товар бракован, при условии что тест дал положительный результат.  
При этом вероятность отсутствия брака при том, что тест показал **-** будет мала. Но для этого случая можно сделать симметричный тест, который сделает большой уже $\mathbb{P}(NB|_)$

#### (3)  
Пусть $p_0$ - вероятность брака, а $p_1$ - ошибка прибора (предполагаем, что ошибки обоих родов равны и равны $p_1$)  
Тогда можем переписать формулу для $\mathbb{P}(B|+)$ в терминах $p_0$ и $p_1$:  
$$
\mathbb{P}(B|+) = \frac{(1 - p_1)p_0}{(1 - p_1)p_0 + p_1(1 - p_0)} = \frac{p_0 - p_0p_1}{p_0 + p_1 - 2p_0p_1} \lt\gt \frac{1}{2} \Leftrightarrow 2p_0 - 2p_0p_1 \lt\gt p_0 + p_1 - 2p_0p_1 \Leftrightarrow p_0 \lt\gt p_1
$$  
Из этого делаем вывод:  
- Тест работает лучше монетки, если $p_0 \gt p_1$, то есть вероятность брака больше вероятности ошибки прибора  
- Тест работает хуже монетки, если $p_0 \lt p_1$, то есть вероятность брака меньше вероятности ошибки прибора  
- Тест работает как монетка, если вероятности равны (что и получилось в пункте 1)
